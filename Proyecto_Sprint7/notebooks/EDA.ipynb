```python
import pandas as pd

# Cargar el archivo CSV (asegrate que est en la carpeta raz del proyecto)
df = pd.read_csv('../vehicles_us.csv')

# Mostrar las primeras filas
df.head()
import plotly.express as px
 
# Crear un histograma de la columna 'price'
fig = px.histogram(df, x='price', nbins=50, title='Distribucin de precios de vehculos')
fig.show()

# Crear un grfico de dispersin de 'model_year' vs 'price' coloreado por 'condition'
fig = px.scatter(df, x='model_year', y='price', color='condition', title='Precio vs. Ao del modelo por condicin')
fig.show()


```
