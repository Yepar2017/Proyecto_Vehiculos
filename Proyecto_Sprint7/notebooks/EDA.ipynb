```python
import pandas as pd

# Cargar el archivo CSV (asegúrate que está en la carpeta raíz del proyecto)
df = pd.read_csv('../vehicles_us.csv')

# Mostrar las primeras filas
df.head()
import plotly.express as px
 
# Crear un histograma de la columna 'price'
fig = px.histogram(df, x='price', nbins=50, title='Distribución de precios de vehículos')
fig.show()

# Crear un gráfico de dispersión de 'model_year' vs 'price' coloreado por 'condition'
fig = px.scatter(df, x='model_year', y='price', color='condition', title='Precio vs. Año del modelo por condición')
fig.show()


```
